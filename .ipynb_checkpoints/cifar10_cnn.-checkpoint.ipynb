# **STEP 1 - LOADING THE DATASET**

```python
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train = X_train.astype('float32') / 255.0   # Normalize to [0, 1]
X_test  = X_test.astype('float32') / 255.0
# One-hot encode labels
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 10)
y_test  = to_categorical(y_test,  10)
```

```python
# Print dataset shapes
print("Training Images Shape:", X_train.shape)
print("Training Labels Shape:", y_train.shape)

print("Testing Images Shape:", X_test.shape)
print("Testing Labels Shape:", y_test.shape)
```

# **STEP 2 — BUILDING THE CNN MODEL**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout
)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(10, activation='softmax')
])

model.summary()
```

The implemented model follows a structured convolutional architecture consisting of two convolutional layers with increasing filter depth (32 and 64 filters), max pooling operations for dimensionality reduction, a fully connected dense layer with 128 neurons, and a dropout layer to improve regularization. Based on the model summary, the architecture contains 315,722 trainable parameters, which provides sufficient representational capacity for feature extraction while remaining computationally efficient for CIFAR-10 image classification. The progressive reduction in feature map dimensions from 30×30 to 6×6 demonstrates effective spatial abstraction, allowing the network to preserve relevant visual information while reducing complexity. The inclusion of Dropout (0.5) after the dense layer is expected to mitigate overfitting and improve generalization performance during evaluation. Overall, the selected architecture satisfies the project requirements and establishes a balanced foundation for achieving competitive classification accuracy, with further performance improvements expected through experimentation involving data augmentation, batch normalization, or transfer learning.

**STEP 3 - COMPILE AND TRAIN**

```python
# =========================
# STEP 3 – COMPILE & TRAIN
# =========================

# Compile the CNN model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=64,
    validation_split=0.2
)

# Display training completion message
print("Model training completed!")
```

```python
# =========================
# SHOW ACCURACY & LOSS GRAPHS
# =========================

import matplotlib.pyplot as plt

# Accuracy graph
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss graph
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

**Discussion / Analysis**

The CNN model was compiled using the Adam optimizer and categorical crossentropy loss function, which are suitable for multi-class image classification problems such as CIFAR-10. The model was trained for 20 epochs with a batch size of 64 and a validation split of 20% to monitor performance during training. Throughout the training process, the model gradually improved its ability to classify images by learning important visual features from the dataset. The generated training and validation graphs help evaluate convergence behavior, detect possible overfitting, and analyze the overall learning performance of the network before proceeding to model evaluation and testing.

**STEP 4**

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", test_accuracy)
```

```python
import numpy as np

# Predict classes
y_pred = model.predict(X_test)

# Convert probabilities to class labels
y_pred_classes = np.argmax(y_pred, axis=1)

# Convert one-hot encoded labels back to single labels
y_true = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

# CIFAR-10 class labels
class_names = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]

# Plot confusion matrix
plt.figure(figsize=(10,8))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=class_names,
    yticklabels=class_names
)

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')

plt.show()
```

## **STEP 5 — EXPERIMENTS**

### **Experiment A: Add Dropout(0.5) after each Dense layer**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout
)

# Redefine the model with an additional Dropout layer after the first Dense layer
model_exp_A = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5), # Existing Dropout

    Dense(64, activation='relu'), # New Dense layer
    Dropout(0.5), # Additional Dropout

    Dense(10, activation='softmax')
])

model_exp_A.summary()
```

The model has been redefined with the additional dropout layer. Now, let's compile and train this new model.

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

# Load and preprocess data (if not already defined)
# This re-initializes X_train, y_train, X_test, y_test if they are missing
(X_train, y_train_raw), (X_test, y_test_raw) = cifar10.load_data()
X_train = X_train.astype('float32') / 255.0
X_test  = X_test.astype('float32') / 255.0
y_train = to_categorical(y_train_raw, 10)
y_test  = to_categorical(y_test_raw,  10)

# Compile the new CNN model
model_exp_A.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history_exp_A = model_exp_A.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=64,
    validation_split=0.2
)

# Display training completion message
print("Model (Experiment A) training completed!")
```

STEP 6 - DEPLOY THE MODEL ON THE WEB


```python
model_exp_A.save("cifar10_cnn_model.keras")

print("Model saved successfully as cifar10_cnn_model.h5")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# CIFAR-10 class labels
class_names = [
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

# Predict test set
y_pred_probs = model_exp_A.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Convert one-hot encoded labels back to integers
y_true = np.argmax(y_test, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10,8))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix - CIFAR-10")
plt.show()

# Classification Report
print("Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))
```

```python
import random

print("Validation Runs:")

for i in range(10):
    index = random.randint(0, len(X_test) - 1)

    sample_image = X_test[index]
    actual_label = class_names[y_true[index]]

    prediction = model_exp_A.predict(np.expand_dims(sample_image, axis=0))
    predicted_label = class_names[np.argmax(prediction)]
    confidence_score = np.max(prediction) * 100

    plt.imshow(sample_image)
    plt.title(f"Actual: {actual_label} | Predicted: {predicted_label}")
    plt.axis('off')
    plt.show()

    print(f"Run {i+1}")
    print(f"Actual Label    : {actual_label}")
    print(f"Predicted Label : {predicted_label}")
    print(f"Confidence Score: {confidence_score:.2f}%")
    print("-" * 50)
```



   The CNN model achieved good classification performance on the CIFAR-10 dataset and was able to correctly identify most image categories during testing and validation. The confusion matrix showed that visually similar classes such as cats and dogs occasionally produced misclassifications, which is common in image recognition tasks. The validation runs demonstrated that the model can predict uploaded images with strong confidence scores, indicating that the trained model generalized effectively on unseen data. Deploying the model through a Streamlit web application also provided an interactive way for users to test the classifier in real time.
